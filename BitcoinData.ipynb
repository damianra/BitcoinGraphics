```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('copy.csv', decimal=',')
df = data.copy()
df
```

```python
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt
```

```python
encoder = LabelEncoder()
df['Fecha'] = encoder.fit_transform(df['Fecha'])
df
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=15 )
plt.yticks(fontsize=15 )
plt.plot(df['Fecha'], df['var.'], 'ro')
plt.ylabel('Variacion %')
plt.xlabel('Fecha')
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=15 )
plt.yticks(fontsize=15 )
plt.plot(df['Último'], df['var.'], 'ro')
plt.ylabel('Variacion %')
plt.xlabel('Cierre')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=15 )
plt.yticks(fontsize=15 )
plt.plot(df['Último'], df['Apertura'], 'ro')
plt.ylabel('Apertura')
plt.xlabel('Cierre')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=15 )
plt.yticks(fontsize=15 )
plt.plot(df['Fecha'], df['Último'], 'ro')
plt.ylabel('Fecha')
plt.xlabel('Cierre')
plt.show()
```
